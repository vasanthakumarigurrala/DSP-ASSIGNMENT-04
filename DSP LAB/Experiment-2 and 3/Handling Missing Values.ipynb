## Fillna,Replace and interpolate functions

#### Code #1: Filling null values with a single value 

```python
df=data
df.fillna(0)
```

```python
df.isnull().sum()
```

#### Code #2: Filling null values with the previous ones 

```python
df1=data
df1.fillna(method="pad")
```

```python
df1.isnull().sum()
```

#### Code #3: Filling null value with the next ones 

```python
df2=data
df2.fillna(method="bfill")
```

```python
df2.isnull().sum()
```

#### Code #4: Filling null values in CSV File 

```python
df3=data
df3["Gender"].fillna("No Gender",inplace= True)
df3
```

```python
df3.isnull().sum()
```

#### Code #5: Filling a null values using replace() method 

```python
df4=data
df4.replace(to_replace=np.nan,value=999999)
```

```python
df4.isnull().sum()
```

#### Code #6: Using interpolate() function to fill the missing values using linear method. 

```python
df5=data
df.interpolate(method ='linear', limit_direction ='forward')
```

```python
df5.isnull().sum()
```

## dropna() function

#### Code #1: Dropping rows with at least 1 null value. 

```python
#Dropping missing values using dropna()
df6=data
df6.dropna() 
#df6.dropna(axis=0)
```

#### Code #2: Dropping rows if all values in that row are missing. 

```python
df7=data
df7.dropna(how="all")
```

#### Code #3: Dropping columns with at least 1 null value. 

```python
df8=data
df8.dropna(axis=1)
```

#### Code #4: Dropping Rows with at least 1 null value in CSV file 

```python
df9=data
df9=df9.dropna(axis=0,how="any")
df9
```
