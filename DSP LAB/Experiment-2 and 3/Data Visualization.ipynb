```python
import matplotlib.pyplot as plt
import numpy as np
country=['usa','India','germany','UK','France']
GDP_Per_Capita=[48000,38000,50000,45000,44000]
plt.bar(country,GDP_Per_Capita)
plt.title("Bar Graph")
plt.xlabel("Country")
plt.ylabel("Capita")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
country=['usa','India','germany','UK','France']
GDP_Per_Capita=[48000,38000,50000,45000,44000]
plt.barh(country,GDP_Per_Capita,color="pink")
plt.title("Bar Graph")
plt.xlabel("Country")
plt.ylabel("Capita")
plt.show()
```
